## $$\text{Drag Coeff. Comparison}$$

```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the different datasets
df1 = pd.read_csv("Data/dragCoeff/coarse2.dat", sep=r"\s+")
df2 = pd.read_csv("Data/dragCoeff/medium.dat", sep=r"\s+")
df3 = pd.read_csv("Data/dragCoeff/fine.dat", sep=r"\s+")

# Coarse mesh
time1 = df1["Time"]
Cd1 = df1["Cd"]

Cd_last1 = Cd1.iloc[-1]
t_last1 = time1.iloc[-1]

# Medium mesh
time2 = df2["Time"]
Cd2 = df2["Cd"]

Cd_last2 = Cd2.iloc[-1]
t_last2 = time2.iloc[-1]

# Fine mesh
time3 = df3["Time"]
Cd3 = df3["Cd"]

Cd_last3 = Cd3.iloc[-1]
t_last3 = time3.iloc[-1]


start = 30

# Creating the figure
fig, ax = plt.subplots(1, 3, figsize=(20, 6), constrained_layout=False)
fig.suptitle("Comparison Between Different Meshes", fontsize=16)

# Plotting the coarse data
ax[0].plot(time1[15:], Cd1[15:], label=rf"$\bar{{C}}$_d")
ax[0].scatter(t_last1, Cd_last1, color='r', marker='x', label=f"Final: {Cd_last1:.3f}")
ax[0].set_title("5748 Cells")
ax[0].set_xlabel("Time")
ax[0].set_ylabel("Cd")
ax[0].grid(True)
ax[0].legend()

# Plotting the medium data
ax[1].plot(time2[start:], Cd2[start:], label=rf"$\bar{{C}}$_d")
ax[1].scatter(t_last2, Cd_last2, color='r', marker='x', label=f"Final: {Cd_last2:.3f}")
ax[1].set_title("13 248 Cells")
ax[1].set_xlabel("Time")
ax[1].set_ylabel("Cd")
ax[1].grid(True)
ax[1].legend()

# Plotting the fine data
ax[2].plot(time3[49:], Cd3[49:], label=rf"$\bar{{C}}$_d")
ax[2].scatter(t_last3, Cd_last3, color='r', marker='x', label=f"Final: {Cd_last3:.3f}")
ax[2].set_title("32 500 Cells")
ax[2].set_xlabel("Time")
ax[2].set_ylabel("Cd")
ax[2].grid(True)

ax[2].legend()

plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.title('Plot over plot for the three different mesh resolutions', fontsize=18)
plt.plot(time1[10:], Cd1[10:], label='5748 cells')
plt.plot(time2[start:], Cd2[start:], label='13 248 cells')
plt.plot(time3[start:], Cd3[start:], label='32 500 cells')
plt.tight_layout()
plt.legend()
plt.show()
```
