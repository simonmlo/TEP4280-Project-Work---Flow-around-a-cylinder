```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ddtScheme Eueler & div(phi, U) Gauss Linear - Reference simulation
EulerGaussLinear = pd.read_csv("Data/numericalDifferences/EulerGaussLinear.dat", sep=r'\s+')
timeEGL = EulerGaussLinear["Time"]
cdEGL   = EulerGaussLinear["Cd"]
cdEGL_avarage = np.mean(cdEGL[400:])

# ddtScheme Euler & div(phi, U) Gauss upwind
EulerGaussUpwind = pd.read_csv("Data/numericalDifferences/EulerGaussUpwind.dat", sep=r'\s+')
timeEGU = EulerGaussUpwind["Time"]
cdEGU   = EulerGaussUpwind["Cd"]
cdEGU_avarage = np.mean(cdEGU[300:])

# ddtScheme Euler & div(phi, U) Gauss linearUpwind grad(U)
EulerLinearUpwind = pd.read_csv("Data/numericalDifferences/EulerLinearUpwind.dat", sep=r'\s+')
timeELU = EulerLinearUpwind["Time"]
cdELU   = EulerLinearUpwind["Cd"]
cdELU_avarage = np.mean(cdELU)

# ddtScheme Euler & div(phi, U) Gauss LUST grad(U)
EulerLUST = pd.read_csv("Data/numericalDifferences/EulerLUST.dat", sep=r'\s+')
timeELUST = EulerLUST["Time"]
cdELUST   = EulerLUST["Cd"]
cdELUST_avarage = np.mean(cdELUST)

# ddtScheme backward & div(phi, U) Gauss linearUpwind grad(U)
BackwardLinearUpwind = pd.read_csv("Data/numericalDifferences/BackwardLinearUpwind.dat", sep=r'\s+')
timeBLU = BackwardLinearUpwind["Time"]
cdBLU   = BackwardLinearUpwind["Cd"]
cdBLU_avarage = np.mean(cdBLU)

# ddtScheme backward & div(phi, U) Gauss upwind
BackwardLUST = pd.read_csv("Data/numericalDifferences/BackwardLUST.dat", sep=r'\s+')
timeBLUST  = BackwardLUST["Time"]
cdBLUST   = BackwardLUST["Cd"]
cdBLUST_avarage = np.mean(cdBLUST)

start = 30

plt.figure(figsize=(12, 6))
plt.suptitle('How different schemes affect the continuation of the simulation', fontsize=18)

# This first plot is for the reference simulation
plt.plot(timeEGL[start:],   cdEGL[start:], linestyle='--', color='black', label=rf'ddtScheme Euler & div(phi, U) Gauss Linear, $\bar{{C}}_d \approx {cdEGL_avarage:.3f}$')

plt.plot(timeEGU,   cdEGU, label=rf'ddtScheme Euler & div(phi, U) Gauss Upwind, $\bar{{C}}_d \approx {cdEGU_avarage:.3f}$')
plt.plot(timeELU,   cdELU, label=rf'ddtScheme Euler & div(phi, U) Gauss linearUpwind grad(U), $\bar{{C}}_d \approx {cdELU_avarage:.3f}$')
plt.plot(timeELUST, cdELUST, label=rf'ddtScheme Euler & div(phi, U) Gauss LUST grad(U), $\bar{{C}}_d \approx {cdELUST_avarage:.3f}$')
plt.plot(timeBLU,   cdBLU, label=rf'ddtScheme backward & div(phi, U) Gauss linearUpwind grad(U), $\bar{{C}}_d \approx {cdBLU_avarage:.3f}$')
plt.plot(timeBLUST, cdBLUST, label=rf'ddtScheme backward & div(phi, U) Gauss LUST grad(U), $\bar{{C}}_d \approx {cdBLUST_avarage:.3f}$')

plt.tight_layout()
plt.legend(loc='lower right')
plt.grid()
plt.show()
```
